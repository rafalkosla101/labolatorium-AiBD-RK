```python
import matplotlib.pyplot as plt
import numpy as np

#Rafa Kola 

def f(x):
    return x**2 + 5


z1=np.arange(-1,1,0.1)
z2=np.arange(-6,6,0.1)
z3=np.arange(0,5,0.1)

plt.plot(z1,f(z1))
plt.title("Wykres funckji x**2 + 5")
plt.xlabel("Wartoci x")
plt.ylabel("Wartoci f(x)")
plt.legend("f(x)")
plt.grid()
plt.show()
plt.plot(z2,f(z2))
plt.title("Wykres funckji x**2 + 5")
plt.xlabel("Wartoci x")
plt.ylabel("Wartoci f(x)")
plt.legend("f(x)")
plt.grid()
plt.show()
plt.plot(z3,f(z3))
plt.title("Wykres funckji x**2 + 5")
plt.xlabel("Wartoci x")
plt.ylabel("Wartoci f(x)")
plt.legend("f(x)")
plt.grid()
plt.show()
```
