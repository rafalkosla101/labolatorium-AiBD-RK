Poniższy kod otwiera plik z danymi oryginalnymi, a następnie wczytuje dane z 2010 roku zgodnie z poleceniem

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

pth = os.getcwd()
pth = pth[:-14] + "/Original data/pom.csv"

df = pd.read_csv(pth,names = ["ID","Dni od zakupu","Marka","Wiek kupującego","Płeć kupującego","Ocena"],header = 0,dtype={'Płeć kupującego' : str})
df = df.astype({"Płeć kupującego": 'string'})
print(df)
```

```python
df2 = df["Dni od zakupu"].value_counts().sort_index()
ax = df2.plot(kind='bar', title ="Ilość klientów która kupiła odkurzacz",figsize=(15,10),legend=True, fontsize=12)
ax.set_ylabel("Ilość zakupów")
ax.set_xlabel("Dni od wykonania zakupu")
ax.legend(["Ilość wystąpień"])
```

Najwięcej odkurzaczy zostało kupionych w niedawno ubiegłych dniach co może znaczyć, że została przeprowadzona akcja promocyjna lub zastosowana została odpowiednia reklama.  Po niskiej ilości kupionych odkurzaczy w bardziej odległych dniach można również wnioskować, że w tym czasie odbyły się jakieś święta lub poprostu sklepy były w tym okresie zamknięte.

```python
df3 = df.dropna()
df3 = df3.astype({'Wiek kupującego':'int32'})
df3 = df3["Wiek kupującego"].value_counts().sort_index()
ax = df3.plot(kind='bar', title ="Ilość klientów która kupiła odkurzacz w rozróżnieniu na wiek",figsize=(15,10),legend=True, fontsize=12)
ax.set_ylabel("Ilość osób w podanym wieku")
ax.set_xlabel("Wiek osób kupujących")
ax.legend(["Ilość wystąpień"])
```

Wiek kupującyh odkurzacze w wojewodztwie pomorskim jest zróżnicowany, jednakże znaczną przewage ma grupa wiekowa od 30 do 50 lat. Osoby w wieku średnim kupują najczęściej odkurzacze dla swoich rodzin. Jedynie niewielka liczba osób starszych w wieku powyżej 60 roku życia oraz osób młodych poniżej 21 roku życia kupuje wcześniej wspomniane produkty.

```python
df4 = df["Marka"].value_counts().sort_index()
ax = df4.plot(kind='bar', title ="Ilość klientów która kupiła odkurzacz poszczególnych firm",figsize=(15,10),legend=True, fontsize=12)
ax.set_ylabel("Ilość zakupów")
ax.set_xlabel("Nazwa Firmy")
ax.legend(["Ilość wystąpień"])
```

Sprzedaż odpowiednich marek sytułuje się na podobnych poziomie w przypadku większości firm, jedynym wyróżnikiem jest firma Tefal, której sprzedaż znacząco przekracza wyniki konkurentów. Z powyższych danych można wnioskować, że firma Tefal prowadzi dobrą reklame lub jakość ich produktów jest wyższa niż pozostałych kunkurentów.

```python
df5 = df["Płeć kupującego"].value_counts().sort_index()
ax = df5.plot(kind='bar', title ="Ilość klientów która kupiła odkurzacz w rozróżnieniu na płeć ",figsize=(15,10),legend=True, fontsize=12)
ax.set_ylabel("Ilość osób")
ax.set_xlabel("Płec osoby kupującej")
ax.legend(["Ilość wystąpień"])
```

Z powyższego wykresu można odczytać dość równowmiery rozkładpłci osób kupujących odkurzacze w województwie pomorskim. Co zawrto w przypadku bardzo dużej grupy osób nie zdefiniowano płci co w przypadku takiego rozkładu nie daje jednoznacznych informacji o kupująych odkurzacze klientach.

```python
df6 = df["Ocena"].value_counts().sort_index()
ax = df6.plot(kind='bar', title ="Oceny klientów kupujących odkurzacze",figsize=(15,10),legend=True, fontsize=12)
ax.set_ylabel("Ilość Ocen")
ax.set_xlabel("Wartość oceny")
ax.legend(["Ilość wystąpień"])
```

Oceny odkurzaczy wachają się dośc mocno. Wiekszość odkurzaczy kupionych w województwie pomorskim otrzymało ocene w zakresie od 1.0 do 3.5 - wynika z tege,że są to odkurzacze średniej jakości. Co jednak warto również zauważyć ilość ocen bardzo pozytywnych(5.0,4.5) jest o wiele większa niż ilość ocen z przedziału od 0 - 0.5

```python
df7 = df.groupby(["Marka"]).mean()
ax = df7['Ocena'].plot(kind='bar', title ="Średnai ocen poszczególnych firm",figsize=(15,10),legend=True, fontsize=12)
ax.set_ylabel("Wartość średniej")
ax.set_xlabel("Nazwa firmy")
```

średnie oceny poszczególnych firm nie różnią sie od siebie znacząco. Tefel ma niezwykle małą przewage oceny średnije nad pozostałymi uczestnikami badania, natomiast Samsung w przypadku zebranych danych wypada najgorzej.
